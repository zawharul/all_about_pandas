```python
import pandas as pd

# Example DataFrame
data = {'animal': ['hen', 'cat', 'duck', 'hen', 'cat', 'duck']}
df = pd.DataFrame(data)

# Create a mapping dictionary
animal_mapping = {'hen': 0, 'cat': 1, 'duck': 2}

# Apply the mapping to the column
df['animal_mapped'] = df['animal'].map(animal_mapping)

print(df)
```
